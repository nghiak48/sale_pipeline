```python
import pyarrow.parquet as pq
import pandas as pd

def inspect_parquet(filepath: str, hour: int):
    df = pq.read_table(filepath).to_pandas()
    
    print(f"Total rows: {len(df)}")
    print(f"sale_date range: {df['sale_date'].min()} → {df['sale_date'].max()}")
    print(f"\nRows per hour:")
    print(df.groupby(df['sale_date'].dt.hour)['sale_id'].count())
    print(f"\nSample data:")
    print(df[['sale_id', 'sale_date', 'amount']].sort_values('sale_date').head(10))

    
```

```python
# Chạy với file hour=01
inspect_parquet(
    "/Users/nghia/sale_pipeline/data_lake/raw/sales/year=2026/month=05/day=01/hour=01/batch_2026-05-01T01.parquet",
    hour=1)
```
